# Setting Up and Authorizing Google Sheets and Google Apps Script

This guide will walk you through the steps needed to get started with Google Sheets and Google Apps Script, including obtaining your Google Voice email and authorizing your script. Follow the steps carefully to ensure that everything is set up properly.

### Step 1: Access the Google Sheet

To get started, you need to either:

- **Make a Copy** of this [Google Sheet](https://docs.google.com/spreadsheets/d/1nHQHiX0R16PqwseTyrbWU5oY2LtopsZov71ZhyxkbOg/edit?gid=931564054#gid=931564054) or

### Step 2: Get a Google Voice Number

To add the text functionality to your project, you will need a Google Voice number. If you don't have one, follow these steps:

1. Go to [Google Voice](https://voice.google.com/u/) and sign in with your personal Google Account.
2. Follow the prompts to select a Google Voice number. Send a text to your personal phone number and reply to the number.

image.png


### Step 3: Retrieve Your Google Voice Email Address

Next, retrieve the email address associated with your Google Voice number. To do this:

1. **Send a Text** from your Google Voice number to your own phone number.
2. **Reply to the Text** from your phone.
3. You will receive an email in your **personal Gmail account** (Note: It must be a personal account as Workspace accounts require a $5/month subscription). This email will contain the address associated with your Google Voice account.


image.png

### Step 4: Store Your Google Voice Email in the Google Sheet

Once you have your Google Voice email:

1. Open the Google Sheet you made a copy of or created in Step 1.
2. Go to the **"Info"** sheet.
3. **Paste** the Google Voice email address into **cell B2**.
image.png

### Step 5: Open Google Apps Script

Now that your sheet is ready, it's time to open the Google Apps Script editor:

1. **Open your Google Sheet**.
2. From the top menu, click on **Extensions** > **Apps Script**.
3. This will open the script editor where you can write and run scripts for your Google Sheet.

image.png


### Step 6: Authorizing the Script

Before you can begin using the Apps Script, you need to authorize it to access your Google Sheet. Heres how you do that:

#### Step 1: Run Code
- Try to run some code in the script.

#### Step 2: Determine Which Google Account Your File Is In
- Make sure you are in the Google Account that the file was copied from. You need to be signed in to the same Google Account that owns the spreadsheet.

You will be asked to review the permissions.

#### Step 3: Trusting the Developer
- The script will ask if you trust the developer of the script. If you wrote it, then you can trust it, but if you're using third-party libraries, you should also trust those as well.

image.png

Click Advanced


image.png

#### Step 4: What the App is Doing
- In this step, Google will display the permissions the script is requesting. In this case, it will be basic permissions to **edit the spreadsheet**.

image.png


#### Step 5: Receive a Security Email from Google
- After authorizing the script, Google will send you a **security email** confirming that you granted access to your Google Account. There is no additional information in that email other than the confirmation of the permissions you granted.

image.png


### Step 7: Begin Coding

Once you have authorized the script, you can start coding. You can now write scripts that interact with the data in your Google Sheet, automate tasks, and integrate other Google services.


### Summary:

- **Access or create** a Google Sheet.
- **Obtain a Google Voice number** and retrieve your Google Voice email by sending and replying to a text.
- **Store your Google Voice email** in cell B2 of the "Info" sheet.
- **Open the Apps Script editor** from the Extensions menu.
- **Authorize** the script by following the steps to grant it permission to access your Google Sheets.

By following these steps, you'll be able to leverage Google Apps Script to automate your workflow with Google Sheets and other Google services.

# Using GenAI in Development

##Prompt Structure **PICO**

To help you remember the key elements when crafting prompts for coding tasks, we've developed the acronym **PICO**. This mnemonic stands for:

1. **P**roblem: What are you trying to solve and why?
2. **I**nputs: What do the inputs or data look like?
3. **C**onstraints: Are there any constraints or nuances to consider?
4. **O**utput: What are you expecting as an output?

## Breakdown of PICO

### 1. **P**roblem: What are you trying to solve and why?
   - **Purpose**: Clearly define the problem you're addressing.
   - **Example**: "I need to create a Python function that sorts a list of dictionaries by a specific key."

### 2. **I**nputs: What do the inputs or data look like?
   - **Details**: Describe the structure, format, and type of input data.
   - **Example**: "The input is a list of dictionaries, each containing 'name' and 'age' keys, like this: `[{'name': 'Alice', 'age': 30}, {'name': 'Bob', 'age': 25}]`."

### 3. **C**onstraints: Are there any constraints or nuances to consider?
   - **Limitations**: Specify any restrictions, requirements, or special conditions.
   - **Example**: "The function should not modify the original list and must handle cases where the key might be missing or `None`."

### 4. **O**utput: What are you expecting as an output?
   - **Result**: Define the desired outcome, including format and structure.
   - **Example**: "The function should return a new list sorted by the 'age' key in ascending order, such as `[{'name': 'Bob', 'age': 25}, {'name': 'Alice', 'age': 30}]`."

## Using PICO: A Step-by-Step Example

### Scenario
You want to create a function that filters and processes user data from a JSON file.

### Applying PICO

1. **P**roblem:
   - "I need to create a JavaScript function that filters out inactive users from a list and calculates their average age."

2. **I**nputs:
   - "The input is an array of user objects, each with 'name', 'age', and 'active' properties, like this:
     ```javascript
     [
       { name: 'Alice', age: 30, active: true },
       { name: 'Bob', age: 25, active: false },
       { name: 'Charlie', age: 35, active: true }
     ]
     ```"

3. **C**onstraints:
   - "The function should only include users where 'active' is `true` and handle cases where 'age' might be missing or not a number."

4. **O**utput:
   - "The function should return an object containing the filtered list of active users and their average age, like this:
     ```javascript
     {
       activeUsers: [
         { name: 'Alice', age: 30, active: true },
         { name: 'Charlie', age: 35, active: true }
       ],
       averageAge: 32.5
     }
     ```"

5. **P**rompt:
   - "Please create a JavaScript function that filters out inactive users from a list, calculates their average age, and handles cases where the 'age' field is missing or not a number. The output should be an object with the filtered users and their average age."


### Final PICO Prompt
"I need to create a JavaScript function that filters out inactive users from a list and calculates their average age. The input is an array of user objects, each with 'name', 'age', and 'active' properties, like this:
```javascript
[
  { name: 'Alice', age: 30, active: true },
  { name: 'Bob', age: 25, active: false },
  { name: 'Charlie', age: 35, active: true }
]
```

### Benefits of Using PICO with Prompt Generation:

- **Clarity**: Ensures all necessary aspects of the problem are addressed.
- **Consistency**: Provides a standardized approach to prompt crafting.
- **Efficiency**: Helps quickly formulate comprehensive and effective prompts.
- **Accuracy**: Reduces the likelihood of missing important details.
- **LLM Optimization**: By adding the "Prompt" step, you can get the best possible phrasing for the prompt to yield the most accurate results from an LLM.
